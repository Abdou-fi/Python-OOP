```python
from abc import ABC, abstractmethod
import math


class Form(ABC):
    @abstractmethod
    def area(self):
        pass


class Square(Form):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side


class Circle(Form):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius
    
class Rectangle(Form):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
    
square1 = Square(5)
print(f"Area of the square1: {square1.area()}")

circle1 = Circle(3)
print(f"Area of the circle1: {circle1.area()}")

rectangle = Rectangle(6, 4)
print(f"Area of the rectangle: {rectangle.area()}")
```
